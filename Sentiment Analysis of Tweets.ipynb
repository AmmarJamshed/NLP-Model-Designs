### 2.4.2. Import classifier Libraries

```python
import nltk
from nltk.corpus import twitter_samples
from nltk.twitter import Twitter
from nltk.classify import NaiveBayesClassifier
```

```python
import nltk
nltk.download('twitter_samples')
```

### 2.4.3. Split the dataset into training and test by negative and positive tweets.

```python
# Load the positive and negative tweet datasets
positive_tweets = twitter_samples.strings('positive_tweets.json')
negative_tweets = twitter_samples.strings('negative_tweets.json')
```

```python
positive_tweets
```

```python
negative_tweets
```

```python
# Split the datasets into training and testing sets
train_pos = positive_tweets[:4000]
test_pos = positive_tweets[4000:]
train_neg = negative_tweets[:4000]
test_neg = negative_tweets[4000:]
```

```python
# Create a combined training dataset
train_data = [(tweet, 'pos') for tweet in train_pos] + [(tweet, 'neg') for tweet in train_neg]
```

```python
# Create a combined testing dataset
test_data = [(tweet, 'pos') for tweet in test_pos] + [(tweet, 'neg') for tweet in test_neg]
```

```python
# Define a function to extract features from a tweet
def extract_features(tweet):
    features = {}
    for word in nltk.word_tokenize(tweet):
        features[word] = True
    return features
```

```python
# Extract the features from the training and testing datasets
train_set = [(extract_features(tweet), label) for (tweet, label) in train_data]
test_set = [(extract_features(tweet), label) for (tweet, label) in test_data]
```

```python
test_set
```

### 2.4.4. Train and test the baseline classifier using NaiveBayesClassifier

```python
# Train a Naive Bayes classifier
classifier = NaiveBayesClassifier.train(train_set)
```

```python
# Print the accuracy of the classifier
print("Accuracy:", nltk.classify.util.accuracy(classifier, test_set))
```

- Our Baseline Accuracy is 97.3% 

```python
# Print the most informative features of the classifier
print(classifier.show_most_informative_features())
```

- Our most important features show that most sentiments are by texts on tweets show positive sentimentality.

## 2.6. Programming Style 

We will test a few functionalities with our data for error cleaning before directly applying them to our model-building process.

```python
# take tweets as a feature
file = open('train_tweets.csv', 'rt')
text = file.read()
file.close()
```

```python
import re

def preprocess_tweet(tweet):
    # Remove hashtags, mentions, URLs, and other unnecessary characters
    tweet = re.sub(r'#\w+', '', tweet)
    tweet = re.sub(r'@\w+', '', tweet)
    tweet = re.sub(r'https?:\/\/\S+', '', tweet)
    tweet = re.sub(r'[^a-zA-Z\s]', '', tweet)
    
    # Lowercase all words
    tweet = tweet.lower()
    
    # Tokenize the tweet and return the list of words
    return tweet
```

```python
preprocess_tweet(text)
```

### 2.6.2 Remove Stop Words from train dataset

```python
import nltk
from nltk.corpus import stopwords

# Get the list of stop words
stop_words = stopwords.words('english')

# Preprocess the tweet and remove stop words
def preprocess_tweet_sw(tweet):
    # Remove hashtags, mentions, URLs, and other unnecessary characters
    tweet = re.sub(r'#\w+', '', tweet)
    tweet = re.sub(r'@\w+', '', tweet)
    tweet = re.sub(r'https?:\/\/\S+', '', tweet)
    tweet = re.sub(r'[^a-zA-Z\s]', '', tweet)
    
    # Lowercase all words
    tweet = tweet.lower()
    
    # Tokenize the tweet and remove stop words
    words = [word for word in nltk.word_tokenize(tweet) if word not in stop_words]
    
    return words
```

```python
preprocess_tweet_sw(text)
```

## 2.9. Classification Approach

- We will test ensemble and tree-based approaches on our dataset to see which one gives the best result.


### 2.9.1 Classification Approach - Random Forest Classifier

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create a CountVectorizer object to create a bag of words representation
vectorizer = CountVectorizer()

# Use the fit_transform method to create a bag of words representation of the text data
X = vectorizer.fit_transform(df['tweet'])

# Extract the labels (i.e., the class labels) from the DataFrame
y = df['label']
```

```python
from sklearn.model_selection import train_test_split

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create a Random Forest classifier
clf = RandomForestClassifier()

# Use the fit method to train the classifier
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Use the predict method to generate predictions for the test set
predictions = clf.predict(X_test)

# Calculate the accuracy of the predictions
accuracy = accuracy_score(y_test, predictions)
```

```python
print("Accuracy: ", accuracy)
```

Thus we can see our accuracy with Random forest classification is 96.07%, which is 1.23% less than our baseline model's performance.

```python
# Try adjusting the hyperparameters of the classifier
clf = RandomForestClassifier(n_estimators=100, max_depth=10)

# Fit the classifier with the adjusted hyperparameters
clf.fit(X_train, y_train)

# Calculate the accuracy of the predictions with the adjusted classifier
predictions = clf.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print("Accuracy: ", accuracy)
```

After Hyperparameter tuning our Random forest classification's accuracy is 93% overall which is 3.07% less than our original score and 4.30% less than our baseline model's performance.

### 2.9.2. Classification Approach - Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

# Create a Decision Tree Classifier
clf = DecisionTreeClassifier()

# Use the fit method to train the classifier
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Use the predict method to generate predictions for the test set
predictions = clf.predict(X_test)

# Calculate the accuracy of the predictions
accuracy = accuracy_score(y_test, predictions)
```

```python
print("Accuracy: ", accuracy)
```
